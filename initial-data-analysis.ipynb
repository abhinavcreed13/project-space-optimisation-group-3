#### Building name with room description for Parkville campus

```python
parkville_uom_space_df = uom_space_df[uom_space_df['Campus Code']=='PAR']
parkville_uom_space_df.shape
```

```python
parkville_uom_space_df.groupby('Room Category').count()
```

##### Join Space and Room Category dataframes

```python
rm_category_type_df['Room Type']=rm_category_type_df['Room Type'].str.strip()
parkville_uom_space_df['Room Type']=parkville_uom_space_df['Room Type'].astype(str).str.strip()
# left outer join
joined_uom_space_df = pd.merge(parkville_uom_space_df,rm_category_type_df,on=['Room Category','Room Type'],how='left')
joined_uom_space_df.shape
```

```python
joined_uom_space_df.groupby('Room Category').count()
```

```python
joined_uom_space_df[joined_uom_space_df['Room Type Definition'].isna()].groupby('Room Type').count()
```

Room Type=100 is not provided in room category data

#### Building name with room description for all campuses 

```python
# left outer join
rm_category_type_df['Room Type']=rm_category_type_df['Room Type'].str.strip()
uom_space_df['Room Type']=uom_space_df['Room Type'].astype(str).str.strip()
joined_uom_space_df = pd.merge(uom_space_df,rm_category_type_df,on=['Room Category','Room Type'],how='left')
joined_uom_space_df.shape
```

```python
joined_uom_space_df.groupby('Room Category').count()
```

```python
joined_uom_space_df[joined_uom_space_df['Room Type Definition'].isna()].groupby('Room Type').count()
```

Only Room type 100 is missing from Room Category Type dataset
